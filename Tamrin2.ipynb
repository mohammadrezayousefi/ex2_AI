```python
import numpy as np
from matplotlib import pyplot
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn import metrics
from sklearn.metrics import r2_score
```

```python
# dataset :
x_1 = np.random.rand(200,1) *8-4
x_1= np.random.rand(200,1) *8-4
x_3=np.random.rand(200,1) *8-4
x_4=np.random.rand(200,1) *8-4
y_4 =  (x_1 **4 + x_1**3 + x_3**2 - x_4 + 5) / 30
y_4 = y_4 + np.random.normal(scale=0.1, size=y_4.shape)
plt.plot(x_1, y_4, 'ro')
```

```python
from sklearn.model_selection import train_test_split

features = x_1
label = y_4
```

```python
features
```

```python
x_train, x_test, y_train, y_test = train_test_split(features,label,test_size=0.2,random_state=101)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(x_train,y_train, color='red', label='Training')
plt.scatter(x_test,y_test, color='g', label='Test')
plt.title("DATA")
plt.legend(loc='best')
```

```python
# DEGREE = 1

poly_features = PolynomialFeatures(degree=1, include_bias=False)
x_1_poly = poly_features.fit_transform(x_train)
reg1_1 = LinearRegression()
reg1_1.fit(x_1_poly, y_train)
```

```python

x_1_vals = x_train
x_1_vals_poly = poly_features.transform(x_1_vals)
y_4_vals = reg1_1.predict(x_1_vals_poly)
plt.suptitle('Degree 1')
plt.ylabel('Output')
plt.xlabel('Input')
plt.scatter(x_test, y_test, color="g")
plt.plot(x_1_vals, y_4_vals, color="b")
plt.show()
```

```python
print(f" MSE : {metrics.mean_squared_error(y_train, y_4_vals)} | R2 : {r2_score(y_train, y_4_vals)}")
```

```python
from sklearn.model_selection import validation_curve
from sklearn.pipeline import make_pipeline
degrees = np.arange(1,21)

model = make_pipeline(PolynomialFeatures(), LinearRegression())

train_scores, validation_scores = validation_curve(
            model, x_1, y_4,
            param_name='polynomialfeatures__degree',
            param_range=degrees)

plt.figure(figsize=(6,4))
plt.plot(degrees, validation_scores.mean(axis=1), lw=2, label='cross-validation')
plt.plot(degrees, train_scores.mean(axis=1), lw=2, label='training')
plt.legend(loc='best')
plt.xlabel('degree of fiting')
plt.ylabel('Variance')
plt.title('Validation Curve')
plt.tight_layout()
```

```python
for key,val in list.items():

    
    model = make_pipeline(PolynomialFeatures(degree=key), LinearRegression())
    from sklearn.model_selection import learning_curve
    train_sizes, train_scores, validation_scores = learning_curve(
        model, x_1 , y_4,
        train_sizes=np.logspace(-1,0,20))

    plt.figure(figsize=(6,4))
    plt.plot(train_sizes, validation_scores.mean(axis=1),lw=2,color='c', label = 'cross-validation')
    plt.plot(train_sizes, train_scores.mean(axis=1),lw=2,color='b', label = 'Training')
    xj = 0,160
    yj= val,val
    plt.plot(xj, yj, lw=1, color='r',ls='--', label="R2 Score")
    plt.ylim(ymin=-.1, ymax=1)
    plt.legend(loc='best')
    plt.xlabel('train samples')
    plt.ylabel('Variance')
    plt.title(f'Curve for degree: {key}')
    plt.tight_layout()
    

plt.show()
```
